# <span style="color:#cc5500;">Daily Weekly Monthly Tasks</span>

The purpose for having a plan in place for daily, weekly, and monthly DBA tasks is to ensure that critical items regarding SQL Server are never over-looked.  Today’s systems can be quite complex, fast moving and rapidly changing.  While many processes can be automated reducing the need for constant attention, things still go wrong and the DBA needs to know what is happening at all times.  There is little excuse when a disk runs out of space, a database goes off-line, or a scheduled job fails and no one knows about it.  Having a thorough process in place helps to reduce omissions and can also act as a very helpful guide when the DBA is out of the office and someone else is tasked with this responsibility.

### <span style="color:#cc5500;">Daily Checks</span>

<u>**Backups**</u>

Verify that the backups are running and completing successfully.  Typically, most DBA’s will either use a self-created stored procedure to perform database backups and then schedule it with a SQL Agent job, or they will make use of a Database Maintenance plan.  Whichever method you use, validate that they are on track.  Investigate any errors such as the backup failure demonstrated here.

<u>**Event Logs**</u>

Use the Log File Viewer to SQL Server Error Logs, Windows Event Logs, Security Logs.  This is a very helpful tool that brings in the most important information into a single user interface that will quickly help the DBA to identify issues that require attention.

<u>**Additional SQL Server Agent Jobs and after hour activities**</u>

It is quite common to have dozens of scheduled jobs that perform a variety of work.  Many times these jobs will have multiple steps that process and move data, update records in tables, and archive old data.  These can come in the form of SSIS packages, SQL Agent jobs, batch and command files, as well as scheduled processes with Windows Task Scheduler.  It is important to understand which jobs are running, what dependencies they may have, and who to notify in the event of a failure.  

<u>**High Availability**</u>

On a daily basis, ensure that all nodes of all clusters are running and that SQL server is running on the expected node or nodes.  Any other HA technologies that you may configure, for example Log Shipping, Replication, and Always On Availability Groups, should be checked daily to ensure that things are synchronized and running normally.

<u>**High level Performance Check**</u>

Performance Monitoring should be a part of any DBA’s work-life and it is important to develop monitoring plan and understand the ‘rhythm’ of the business.  There is a lot of guidance and documentation on the Web on how to perform monitoring, and what items to monitor, and when to monitor.   The purpose of the High Level Performance check is to get a quick look at what is happening.  Use tools such as Windows Performance Monitor (perfmon) SCCM, Idera, Solarwinds or whatever software you have at your disposal to look at the overall health of the servers.  Most can be confgured to monitor specified items and then send alerts based upon the configured thresholds.   

**<u>Storage</u>**

Monitor and observe all SAN LUN’, DAS, NAS, or whatever drives that contain SQL Server Database files.  Ensure that all system and user database data files and transaction logs have sufficient space.  Also, many SSIS packages may consume or export flat files using a shared file directory so it is important to look at those as well.  Backup disk drives and tape libraries also need to be checked to make sure that they are not running out of space and that the older files are being archived per your archive policy.  Configure alerts to notify the DBA and an alternate staff member of low disk space warnings.

**<u>Make Corrective Actions</u>**

If an issue is identified make corrective actions on a timely basis.  Very rarely do issues work themselves out unattended.  If you see that there is a database corruption issue address it, if there are warnings of failed logins verify that the passwords are correct.  It is critically important to know how to identify that there is a problem, where the problem is, what the impact is, and then how to fix it.

### <span style="color:#cc5500;">Weekly Checks</span>

**<u>User Accounts</u>**

Validate that Domain User accounts and SQL logins that have both server access and database access are in fact, needed.   Most organizations have some amount of turn-over and personnel changes on a periodic basis.   When a person leaves, or when a domain account or login is no longer needed, it should be removed.  Also, applications that connect to SQL Server will connect using an account.  When applications are no longer used on a particular server, remove those accounts as well.  Another way to lower your overall surface area from security breaches and to have better security control, make use of database roles, add individual accounts to the role and then control permissions at the ‘role’ level.

**<u>Database Maintenance and Fragmentation</u>**

It is important to perform maintenance on SQL Server databases.  Common operations include making backups as mentioned earlier, but also to periodically check for and correct fragmentation when it occurs.  Other sections of this notebook cover fragmentation, and what fragmentation is, how to identify it and finally how to fix fragmented indexes using a variety of techniques of either rebuilding or reorganizing them.  Other tasks that will improve performance would be to ensure that each table has updated statistics and that the ‘auto shrink’ and ‘auto close’ options are not set.   Database maintenance also includes checking database integrity and cleaning up maintenance history so that the msdb tables do not get too large.

**<u>Validation of backups and tape archives</u>**

Typically, most customers will back-up their SQL Server databases to hard disk and then have a process in place to copy them to cold disk storage or tape.  These tapes are usually sent off-site for long term archiving.  Develop a plan to periodically recall tapes from storage and test that the backups are valid, that the correct tapes are actually able to be retrieved from the off-site location, loaded, and that the files can be actually be restored.  One way to do this is to use the RESTORE FILELISTONLY command to return a result set containing a list of the database and log files contained in the backup set.  Use RESTORE VERIFYONLY to verify the backup.  This will not restore the backup, it checks to see that the backup set is complete and the entire backup is readable.  However RESTORE VERIFYONLY does not attempt to verify the structure of the data contained in the backup volumes, but it has been enhanced to do additional checking on the data to increase the probability of detecting errors. The goal is to be as close to an actual restore operation as practical.

**<u>Alerting</u>**

Configure alerts to notify the DBA of any events that are important to be aware of should they occur.  There are many alerts that can help a DBA mitigate an issue before it becomes a problem or noticed by end users.  For example, a SQL Server Databases alert can be configured to monitor the % Transaction Log Used counter.  If the transaction log of a monitored database reaches a certain threshold value, the Alert can fire, a command to perform a backup on the log can automatically be programmed and executed (freeing up space in the log), and then the alert can automatically be routed to the appropriate responder.  

**<u>Commit to Learning</u>**

Each week, make a commitment to learning something new.  SQL Server is a very robust, enterprise platform that takes DBA’s years to master.  The amount of information to consume and understand can sometimes be overwhelming and it can be helpful to break things down into smaller buckets, like “Storage”, “Query Tuning”, “Trouble-shooting”, “Reporting Services Architecture”, “T-SQL Syntax”, Transforming data with SSIS”.  Explore the internet for user groups, blogs, wiki’s and sites that can provide a wealth of information. 

**<u>Performance Dashboard</u>**

SQL Server Management Studio has a wealth of helpful reports available to DBA’s out of the box.  Make use of these reports on a weekly basis.   Some of the list of available reports are:

- Disk Usage by Table
- Backup and Restore Events
- Index Usage Statistics…..and many more 

<u>**Plan for capacity**</u>

Create a SQL Script that will capture information about disk usage, percent free space, percent used space, average memory usage, average cpu usage etc.  There are a variety of ways to do this and DBA’s generally will load this data into a private DBA database table or into an Excel workbook.  Each week append the new data into the table or spread sheet.  Once this data is stored on record, this will create opportunities to interact with this data in order to provide ‘trending’ information.   Once you can see the actual trends of your business, you can do true capacity planning based on the facts of real data.

### <span style="color:#cc5500;">Monthly Checks</span>

<u>**Update Documentation**</u>

Create a process to store vital information about the TGH environment.  For example:

- Server Names
- Server and Database Account Names and passwords, (these should not be available in the ‘clear’)
- SQL Instances and the purpose they serve
- Vendor contact information
- Service accounts and gMSA accounts

At least once a month, plan to review for accuracy and correct any changes that are necessary.

**<u>Database Backup and Restore scripts</u>**  

It is critical that the DBA pre-plan several disaster scenarios related to the various types of database restorations that could possibly need to be performed, and then have the pre-built RESTORE DATABASE scripts immediately available.  Most DBA’s have a folder on their PC’s desktop, or in the My Documents folder that contain these scripts.  Another tool available to DBA’s is to make use of SQL Server Management Studio “Solution Explorer”   Create multiple types of restore scripts in your “solution” and the DBA will never have to scramble to find it in an emergency.

It is also important to actually test restoring a production database to DEV environment.  Submit a request and recall a backup from the off-site tape location.   Copy a backup from disk and actually go through the steps of restore.  Without performing an occasional restore, there is no way to know if your backup and DR strategy will actually work in an emergency.

**<u>Look for out-dated databases to archive</u>**

Many DBA’s have a strategy to clean up and remove databases that are not really used anymore but yet at the same time, not quite ready to be deleted.  These would be for databases that are in an “interim” stage.  The DBA may not know if a database can be deleted, or perhaps that database may be needed in a couple of months but you are not sure, yet at the same time, it is serving no current purpose and taking up disk space and server resources.  Consider setting up a separate SQL instance on a minimally performing server for archiving.  Shrink the data files, shrink the log file, attach the database to the archive server and set the database to be Offline, Read Only, or Restricted Use Only.  This way, you are conserving resources on the production servers yet you can quickly make the database available if requested.

**<u>Operating System and SQL Server Updates</u>**

Develop a strategy and documented guidance on when and how to apply Windows Updates, Windows Service Packs and SQL Server Updates and SQL Server Services Packs and Critical Updates (CU’s) to a DEV environment, and then how to deploy them to the production environment.   Understand how to patch a stand-alone server and a clustered server and what potential impact it will have on availability and end users.   

**<u>Improvements</u>**

The DBA should always look for opportunities to make improvements to the environment.

- Do I have a good set of Index Rebuilding scripts or can I make them better?
- The last time I set up Transaction Log shipping, it took me 4 hours.  Can I find a way to reduce this time?
- I found out that there was a SQL Server out there that I didn’t know about for 3 months.  What can I do to ensure that doesn’t happen again?

Think of ways to improve and streamline processes, and develop professional relationships with other DBA’s to find out what they do and how they do it.

**<u>Emergency Contact information</u>**

Make sure that everyone knows how to reach the DBA and that if he or she will be unavailable for an extended period of time, that there is person responsible.  This information should be updated and distributed on a monthly basis.
